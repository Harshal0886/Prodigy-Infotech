# 1. Comparing different Country's mean working hours for all years.

```python
# STEP1: Group by Country for different year and taking mean working hours.
```

```python
group_by_country = data.groupby(['Entity'])['Average annual working hours per worker'].mean().reset_index()
group_by_country_desc = group_by_country.sort_values(by='Average annual working hours per worker', ascending=False)
print(group_by_country_desc)
```

# STEP2:Visualising a appealing bar chart using libraries matplotlib and seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='Entity', y='Average annual working hours per worker', data=group_by_country_desc)
ax.set(xlabel='Country', ylabel='Average Annual Working Hours per Worker', title='Average Annual Working Hours per Worker by Entity')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()

plt.show()
```

# STEP3: Visualising a bar chart in which india position could be seen easily among all countries.

```python
highlight_color = ['skyblue' if country != 'India' else 'salmon' for country in group_by_country_desc['Entity']]

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='Entity', y='Average annual working hours per worker', data=group_by_country_desc, palette=highlight_color)
ax.set(xlabel='Country', ylabel='Average Annual Working Hours per Worker', title='Average Annual Working Hours per Worker by Entity')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()

plt.show()
```

CONCLUSION:

1. South korea has highest mean Annual working hours per worker.
   
2.Trinidad and lobargo has least mean Annual working hours per worker.

3.India ranks 21 for mean Annual working hours per worker.

# 2. Comparing differnt country's working hours for most recent year available..

# STEP1: Group by Country for most recent year available (2017) and taking working hours.

```python
country_2017 = data.groupby('Entity').last().reset_index()
print(country_2017[['Entity', 'Average annual working hours per worker']])
```

# STEP2:Visualising the country's annual working hours for the most recent year 2017

```python
sns.set(style="whitegrid", rc={"axes.facecolor": "#f3f3f3", "grid.color": "#e1e1e1"})

plt.figure(figsize=(14, 8))
ax = sns.barplot(x='Entity', y='Average annual working hours per worker', data=country_2017, palette='rainbow')
ax.set(xlabel='Entity', ylabel='Average Annual Working Hours per Worker (2017)', title='Average Annual Working Hours per Worker by Country in 2017')

ax.yaxis.grid(True)

plt.xticks(rotation=45, ha='right') 
plt.tight_layout()

plt.show()
```

# STEP3: Visualising a bar chart in which india position could be seen easily among all countries for year 2017.

```python
highlight_color = ['lightpink' if country != 'India' else 'salmon' for country in country_2017['Entity']]

sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='Entity', y='Average annual working hours per worker', data=country_2017, palette=highlight_color)
ax.set(xlabel='Country', ylabel='Average Annual Working Hours per Worker', title='Average Annual Working Hours per Worker by Entity')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()

plt.show()
```

CONCLUSION

1.For year 2017 Argentina has highest annual working hours per worker.

2.For year 2017 Vietnam has lowest annual working hours per worker.

3.For year 2017 India ranks 28 for annual working hours per worker

# 3. Comparing within Country how the woking hours changed by different years.

# STEP1:Comparing annual working hours for india through the years .

```python
India_data = data[data['Entity'] == 'India']
print(India_data)
```

# STEP2: Visualising Annual working hours for India past years .

```python
sns.set(style="whitegrid", rc={"axes.facecolor": "#f3f3f3", "grid.color": "#e1e1e1"})

plt.figure(figsize=(14, 8))
ax = sns.barplot(x='Year', y='Average annual working hours per worker', data=India_data, palette='Set2')

ax.set(xlabel='Year', ylabel='Average Annual Working Hours per Worker', title='Average Annual Working Hours per Worker in India (All Years)')

ax.yaxis.grid(True)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

CONCLUSION

1.IN year 1970 annual working hours per worker was 2077.

2.In year 2017 annual working hours per worker was 2117.

3.We can easily observe that in India annual working hours per worker increased from 1970 to 2017

# 4. Comparing top 5 countries for past four years annual working hours.

# STEP1: Get the top 5 countries with the highest average annual working hours for past four years.

```python
recent_years_data = data[ data['Year'] >=  data['Year'].max() - 4]

top_countries = recent_years_data.groupby('Entity')['Average annual working hours per worker'].mean().sort_values(ascending=False).head(5).index

top_countries_data = recent_years_data[recent_years_data['Entity'].isin(top_countries)]
print(top_countries_data)
```

# STEP2: Visualising the top 5 countries for past 4 years years wise.

```python
sns.set(style="whitegrid", rc={"axes.facecolor": "#f3f3f3", "grid.color": "#e1e1e1"})

plt.figure(figsize=(14, 8))
ax = sns.barplot(x='Year', y='Average annual working hours per worker', hue='Entity', data=top_countries_data, palette='Set3')

ax.set(xlabel='Year', ylabel='Average Annual Working Hours per Worker', title='Top 5 Countries with Highest Average Annual Working Hours (Recent 5 Years)')

ax.yaxis.grid(True)

plt.xticks(rotation=45, ha='right')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  


plt.tight_layout(rect=[0, 0, 0.85, 1]) 
plt.show()
```

# CONCLUSION:

1.Top five leading countries for past five years in Annual working hours are Cambodia Malesia Mexico Myanmar Singapore
